```python
# Import Section
import GEOparse
```

```python
# Load Gene expression
gse = GEOparse.get_GEO(filepath="../DataSets/GSE63060_family.soft.gz")
```

```python
gpl_lookup = []

for gpl_name, gpl in gse.gpls.items():
    gpl_lookup = gpl.table.filter(["ID", "ILMN_Gene"]).set_index("ID")[
        "ILMN_Gene"].to_dict()

sampleId = "GSM1539080"
gsm = gse.gsms[sampleId]
gene_names = (gsm.table["ID_REF"]).map(lambda idref: gpl_lookup[idref])

dataset = gsm.table[["ID_REF", "VALUE"]].copy(deep=True)
dataset["GENE"] = gene_names

pivot_table = dataset.pivot_table(index=["GENE", "ID_REF"], columns=[], values="VALUE")
pivot_table.head()
```

```python
from pivottablejs import pivot_ui

pivot_ui(pivot_table)
```
