```python
import pandas as pd
import GEOparse
from sklearn import preprocessing
```

```python
dataset = pd.read_csv("./output.csv", low_memory=False)
```

```python
columns_to_remove = []
for column in dataset.columns:
    if "Unnamed" in column:
        columns_to_remove.append(column)

dataset.drop(columns_to_remove, inplace=True, axis=1)

dataset.columns
```

```python
columns_to_split = []

for column in dataset.columns:
    if "||" in column:
        columns_to_split.append(column)

for column_to_split in columns_to_split:
    splits = column_to_split.split(" || ")
    for split in splits:
        dataset[split] = dataset[column_to_split]

dataset.drop(columns_to_split, inplace=True, axis=1)
```

```python
dataset.to_csv("output.csv", index=False)
```
